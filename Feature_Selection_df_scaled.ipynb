```python
import pandas as pd
import numpy as np
```

```python
file_path = "Newsbooks_Intermediary_Data/res.csv"
res = pd.read_csv(file_path)
```

```python
def eda_fn(x):
    uniq_lst = pd.unique(x).tolist()
    n = len(uniq_lst)
    return n, uniq_lst
```

```python
pd.set_option('display.max_rows', 100)
res_info = res.apply(lambda x: eda_fn(x)).transpose()
res_info['dtype'] = res.dtypes
res_info.reset_index(inplace=True)
res_info.columns = ['feature', 'nunique', 'unique values', 'dtype']
res_info
```

```python
numerical_var_lst = res_info[
                               (res_info['feature'] != 'Ideology')
                               & (res_info['dtype'] != "object") 
                               & (res_info['nunique'] > 1)
                              ]['feature'].tolist()
numerical_var_lst
```

```python
# the last four features are TSNE/PCA, we get rid of those before feature selection
numerical_var_lst_drop_TSNE_PCA = numerical_var_lst[:-4]
len(numerical_var_lst_drop_TSNE_PCA), numerical_var_lst_drop_TSNE_PCA
```

```python
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
scaler.fit(res[numerical_var_lst_drop_TSNE_PCA])

scaled_data = scaler.transform(res[numerical_var_lst_drop_TSNE_PCA])
df_scaled = pd.DataFrame(scaled_data, columns=numerical_var_lst_drop_TSNE_PCA)
```

```python
df_scaled[df_scaled.isna().any(axis=1)]
df_scaled = df_scaled.fillna(0)
df_scaled[df_scaled.isna().any(axis=1)]
```
